```python
# Kütüphanelerimizi import ettim
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv( "animation_movies_enriched_1878_2029.csv")#Veriyi okuma
df.head(10)
```

```python
df.info()
```

```python
# Release_Date sütununu datetime yaptım
df[ "Release_Date"]=pd.to_datetime(df[ "Release_Date"],errors= "coerce")
```

```python
df.describe().T
```

```python
#Eksik veri kontrolü
df.isnull().sum()
```

```python
#metin sütunlarını Unknown ile doldurdum
metın_sutunlari=[ "IMDb_ID", "Studio", "All_Production_Companies", "Director", "Country_Origin", "Animation_Style", "Overview",
                   "Voice_Cast", "Collection_ID", "Poster_URL", "Backdrop_URL", "TMDB_URL", "TMDB_Rating"]
df[metın_sutunlari]=df[metın_sutunlari].fillna( "Unknown")
```

```python
#Sayısal sütunları 0 ile doldurdum
df[ "Movie_Length_Minutes"]=df[ "Movie_Length_Minutes"].fillna(0)
df[ "Release_Year"]=df[ "Release_Year"].fillna(0)
df[ "TMDB_Vote_Count"]=df[ "TMDB_Vote_Count"].fillna(0)
df[ "TMDB_Popularity"]=df[ "TMDB_Popularity"].fillna(0)
```

```python
#Profitability_ROI sütununu ortadan kaldırdım ( baya veri eksikti ,yerine kendim ROI sütunu ekledim)
if  "Profitability_ROI" in df.columns:
    df= df.drop(columns= ["Profitability_ROI"])
```

```python
df_ticari=df[(df[ "Budget_Million_USD"]>0) & (df["Box_Office_Million_USD"]>0)].copy()  
print(f"Ticari Analiz için hazır film sayısı: {len(df_ticari)}")
```

```python
# Yatırım getirisi sütununu oluşturdum
df_ticari[ "ROI"]=df[ "Box_Office_Million_USD"]/df[ "Budget_Million_USD"]
```

```python
# Hangi yılın en çok yatırım getirisi sağladığını buldum
analiz_1=df_ticari[df_ticari[ "Release_Year"] > 0]
karli_yıllar=analiz_1.groupby( "Release_Year")[ "ROI"].mean().sort_values(ascending=False).head(5).reset_index()
karli_yıllar.style.highlight_max(subset= "ROI",color= "lightgreen")
```

```python
# INSIGHT: Disney pazarın %40'ından fazlasını tek başına domine etsede Rus ekolü ile rekabet devam ediyor
analiz_2= df[df[ "Studio"]!= "Unknown"]
s=analiz_2[ "Studio"].value_counts().head(5)
s
```

```python
plt.figure(figsize=(6,3))
s.plot(kind= "bar",color= "blue",title= "Pazarı Ele Geçiren Stüdyolar")
plt.xticks(rotation=45,ha= "right")
plt.show()
```

```python
# INSIGHT:Finansal açıdan 2D animasyonlar hala en güvenli liman olarak karşımıza çıkıyor
# INSIGHT: 3D ler daha yüksek hasılat yapsada yüksek bütçeler yüzünden 2D'nin arkasında kalmıştır
analiz_3=df_ticari.groupby( "Animation_Style")[[ "Box_Office_Million_USD", "ROI"]].mean().sort_values(by= "ROI",ascending=False)
analiz_3
```

```python
hedef_yil=1942
yildizlar=df_ticari[df_ticari[ "Release_Year"]== hedef_yil]["Movie_Name"].unique()
print(yildizlar)
```

### **INSIGHT**: Verilerimizdeki 310.9 ROI değeriyle Bambi, projenin 'Altın Standardı'dır. Bu başarı, yüksek bütçeli teknolojinin her zaman yüksek kârlılık anlamına gelmediğinin, asıl başarının maliyet/kazanç oranında gizli olduğunun en somut kanıtıdır.

```python
analiz_4=df_ticari[df_ticari[ "Country_Origin"]!= "Unknown"]
country_roi=analiz_4.groupby("Country_Origin")[ "ROI"].mean().sort_values(ascending=False)
country_roi
```

### **INSIGHT**: Ülke bazlı analiz yaptım ve analiz sonucu ÇİN ve KORE'nin  ROI'yi 18 katına katladığını görüyorum.

```python
country_roi.head(5).plot(kind= "barh",color= "skyblue",figsize=(9,5))
plt.title( "En Yüksek ROI Oranına Sahip Ülkeler")
plt.ylabel( "Ülkeler")
plt.xlabel( "Ortalama ROI")
plt.show()
```

```python
cin_kore_film=analiz_4[analiz_4[ "Country_Origin"]== "China|South Korea"]
print(cin_kore_film[["Movie_Name", "ROI", "Release_Year"]])
```

```python
# 2016'daki diğer animasyonlara baktım.
df_2016=df_ticari[df_ticari[ "Release_Year"]== 2016].copy().head(10)
df_2016
```

```python
plt.figure(figsize=(6,3))
sns.scatterplot(data=df_2016,x= "Budget_Million_USD",y= "Box_Office_Million_USD",hue= "ROI",size= "ROI",palette= "flare")
plt.title( "2016 Yılı: Bütçe Ve Hasılat Efendileri",fontsize=15)
plt.show()
```

### **INSIGHT**: Grafiğin sol alt köşesindeki MOR nokta düşük bütçeyle devasa bir kar oranına imza atmıştır,sağ üstteki nokta ise yüksek bütçeyle güzel para kazanmış ve 2016'ın Hasılat şampiyonu olmuş.

```python
analiz_5=df_ticari[(df_ticari[ "Director"]!= "Unknown") & (df_ticari[ "TMDB_Rating"]!= "Unknown")]
dır_avg=analiz_5.groupby( "Director")[ "TMDB_Rating"].agg([ "mean", "count"])
dır_avg_sorted=dır_avg.sort_values(by= "mean",ascending=False)
dır_avg_sorted.head(10)
```

```python
dır_avg_sorted.head(10)[ "mean"].plot( kind="barh",title= "En Yüksek Puanlı Top 10 Yönetmen",
                             figsize=(9,5),color= "red")
plt.xlabel("Director",fontsize=12)
plt.ylabel( "TMDB Puanı",fontsize=12)
plt.show()
```

```python
# Düşük bütçeyle yüksek yatırım getirisi getiren filmleri buldum ve BAMBİ yine göz bebeği olarak karşıma çıkıyor.Bu analizden şunu anlıyoruz düşük bütçeylede güzel yatırım getirisi sağlanabilir.
analiz_6=df_ticari[(df_ticari[ "Budget_Million_USD"] <50) & (df_ticari[ "ROI"]>2)]
analiz_6_top=analiz_6[[ "Movie_Name", "Budget_Million_USD", "ROI"]].sort_values(by= "ROI",ascending=False)
print(analiz_6_top.head(7))
```

```python
plt.figure(figsize=(9,6))
sns.barplot(x= "Movie_Name",y= "ROI",hue= "Budget_Million_USD",data=analiz_6_top.head(7))
plt.title( "Düşük Bütçeyle Dev Başarı Elde Eden Filmler")
plt.xticks(rotation=45,ha= "right")
plt.show()
```

### **FINAL INSIGHT**:Animasyon sektörü artık sadece en iyi görseli sunanların değil; düşük bütçeyle (Low Budget) yüksek duygusal bağ ve ROI kurabilenlerin oyun alanı haline geliyor. Verilerimiz gösteriyor ki; devasa bütçeli bir 3D blockbuster projesi yerine, doğru hedef kitleye hitap eden stratejik bir 2D veya bölgesel yapım, yatırımcısına çok daha büyük bir finansal özgürlük alanı tanıyor.
